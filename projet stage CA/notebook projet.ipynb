## 1. Importation des bases de données

```python
import pandas as pd

# Charger les fichiers CSV
# Chargement des données CSV
carac = pd.read_csv("carcteristiques-2022.csv", sep=';', encoding='utf-8')
lieux = pd.read_csv("lieux-2022.csv", sep=';', encoding='utf-8')
vehic = pd.read_csv("vehicules-2022.csv", sep=';', encoding='utf-8')
usagers = pd.read_csv("usagers-2022.csv", sep=';', encoding='utf-8')

# Afficher les 5 premières lignes de chaque table
print("Table: caractéristiques", carac.shape)
display(carac.head())

print("Table: lieux", lieux.shape)
display(lieux.head())

print("Table: usagers", usagers.shape)
display(usagers.head(10))

print("Table: véhicules", vehic.shape)
display(vehic.head(10))
```

Première remarque/modification à effectuer : dans la table caractéristiques la description fournie par le fichier en piece jointe indique la presence d'une colonne 'Num_Acc' au lieu de 'Accident_Id'. On procède à la renommer

```python
carac.rename(columns={"Accident_Id": "Num_Acc"}, inplace=True)
display(carac.head())
```

Fusion des jeux de données : on va créer un dataframe principal qui regroupe toutes les informations pertinentes en une seule table exploitable pour la modélisation. On procède par étapes :

Fusionner usagers et vehicules sur Num_Acc et l'identifiant véhicule (id_vehicule ou num_veh). Cela associe à chaque usager les informations du véhicule impliqué.

Fusionner le résultat avec carac (sur Num_Acc) pour ajouter les caractéristiques de l'accident.

Fusionner enfin avec lieux (sur Num_Acc) pour ajouter les informations de lieu.

```python
# Fusion usagers + véhicules
usagers_veh = pd.merge(usagers, vehic, on=["Num_Acc", "id_vehicule"], how="inner")

# Fusion avec caractéristiques de l'accident
usagers_veh_carac = pd.merge(usagers_veh, carac, on="Num_Acc", how="inner")

# Fusion avec lieux de l'accident
df = pd.merge(usagers_veh_carac, lieux, on="Num_Acc", how="inner")

print("Données fusionnées:", df.shape)
# Afficher un extrait des premières lignes
df.head(10)
```

### 3.2 Encodage des variables catégorielles

La grande majorité des variables explicatives sont **catégorielles** (représentées souvent par des codes numériques dans les données brutes). Pour que les algorithmes de machine learning les exploitent correctement, il faut les **encoder** en variables numériques appropriées :

- Pour les algorithmes basés sur des arbres (Random Forest, XGBoost), il est possible de garder les codes numériques tels quels en s'assurant qu'ils sont bien considérés comme des catégories ordonnées arbitrairement. Cependant, il est plus sûr de réaliser un **one-hot encoding** (encodage binaire) afin de ne pas introduire d'ordre artificiel.
- Pour les algorithmes linéaires (Régression logistique) ou d'autres modèles, un encodage one-hot est généralement nécessaire.
- Alternativement, un `LabelEncoder` peut transformer des catégories en entiers, mais cela serait interprété comme ordinal par la régression logistique, donc préférable de one-hot dans ce cas.

Dans notre approche, nous allons **One-Hot encoder** toutes les variables catégorielles pertinentes. Avant cela, nous pouvons simplifier certains champs :

- **Temps (heure/minute)** : La colonne `hrmn` indique l'heure et minute de l'accident (format HHMM). Plutôt que de traiter cette valeur comme numérique continue, on peut l'exploiter pour créer une variable catégorielle "heure de la journée". Par exemple, on peut extraire l'heure (HH) et en faire une catégorie de 0 à 23.

- **Date (mois)** : De même, on peut utiliser le mois (`mois`) comme catégorie 1-12.

- **Coordonnées GPS (latitude/longitude)** : ces deux variables peuvent être complexes à utiliser directement. Une approche avancée consiste à regrouper les accidents en zones géographiques via clustering K-means, mais pour simplifier, on pourrait soit les exclure de la modélisation, soit arrondir la position à des *bins* (grille) ou regrouper par département/région si cette info est présente.

- **Année de l'accident (`an`)** : si présente, elle est constante (2022 dans tous les cas), donc **sans variance**. On peut la supprimer car elle n'apporte aucune information discriminante.

- **Année de naissance (`an_nais`)** : on a déjà calculé l'`age`. On peut donc se servir de l'`age` comme feature et éventuellement supprimer `an_nais` pour éviter redondance.

- **Identifiants** : On supprime les identifiants uniques qui ne servent pas de prédicteurs (`Num_Acc`, `id_vehicule`, `id_usager`, etc.).

**Appliquons ces transformations :**


```python
# Transformation de l'heure 'hrmn' en catégorie heure (0-23)
if 'hrmn' in df.columns:
    df['heure'] = pd.to_datetime(df['hrmn'], format='%H:%M', errors='coerce').dt.hour # Extraire l'heure (HH) à partir du format "HH:MM"
    df.drop(columns='hrmn', inplace=True)

# Vérification de 'an' (année de l'accident) et suppression s'il existe
if 'an' in df.columns:
    df.drop('an', axis=1, inplace=True)

# Création de l'âge à partir de l'année de naissance, si pertinent
if 'an_nais' in df.columns:
    df.drop('an_nais', axis=1, inplace=True)

# Gestion des coordonnées géographiques (lat, long)
if 'lat' in df.columns and 'long' in df.columns:
    df.drop(['lat','long'], axis=1, inplace=True)  # on les retire pour simplifier 

# Suppression des identifiants uniques
cols_id = [col for col in ['Num_Acc','id_vehicule','id_usager','num_veh'] if col in df.columns]
df.drop(columns=cols_id, inplace=True)
```
