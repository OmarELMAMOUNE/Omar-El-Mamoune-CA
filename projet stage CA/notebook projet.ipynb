## 1. Importation des bases de données

```python
import pandas as pd

# Charger les fichiers CSV
# Chargement des données CSV
carac = pd.read_csv("carcteristiques-2022.csv", sep=';', encoding='utf-8')
lieux = pd.read_csv("lieux-2022.csv", sep=';', encoding='utf-8')
vehic = pd.read_csv("vehicules-2022.csv", sep=';', encoding='utf-8')
usagers = pd.read_csv("usagers-2022.csv", sep=';', encoding='utf-8')

# Afficher les 5 premières lignes de chaque table
print("Table: caractéristiques", carac.shape)
display(carac.head())

print("Table: lieux", lieux.shape)
display(lieux.head())

print("Table: usagers", usagers.shape)
display(usagers.head(10))

print("Table: véhicules", vehic.shape)
display(vehic.head(10))
```

Première remarque/modification à effectuer : dans la table caractéristiques la description fournie par le fichier en piece jointe indique la presence d'une colonne 'Num_Acc' au lieu de 'Accident_Id'. On procède à la renommer

```python
carac.rename(columns={"Accident_Id": "Num_Acc"}, inplace=True)
display(carac.head())
```

Fusion des jeux de données : on va créer un dataframe principal qui regroupe toutes les informations pertinentes en une seule table exploitable pour la modélisation. On procède par étapes :

Fusionner usagers et vehicules sur Num_Acc et l'identifiant véhicule (id_vehicule ou num_veh). Cela associe à chaque usager les informations du véhicule impliqué.

Fusionner le résultat avec carac (sur Num_Acc) pour ajouter les caractéristiques de l'accident.

Fusionner enfin avec lieux (sur Num_Acc) pour ajouter les informations de lieu.

```python
# Fusion usagers + véhicules
usagers_veh = pd.merge(usagers, vehic, on=["Num_Acc", "id_vehicule"], how="inner")

# Fusion avec caractéristiques de l'accident
usagers_veh_carac = pd.merge(usagers_veh, carac, on="Num_Acc", how="inner")

# Fusion avec lieux de l'accident
df = pd.merge(usagers_veh_carac, lieux, on="Num_Acc", how="inner")

print("Données fusionnées:", df.shape)
# Afficher un extrait des premières lignes
df.head(10)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Fonction Cramér’s V
def cramers_v(x, y):
    cm = pd.crosstab(x, y)
    chi2 = chi2_contingency(cm)[0]
    n = cm.values.sum()
    phi2 = chi2 / n
    r, k = cm.shape
    return np.sqrt(phi2 / min(k-1, r-1))

# 2. Variables à comparer à 'grav'
vars_cat = ['sexe', 'catu', 'catv', 'trajet', 'secu1', 'secu2', 'secu3', 'surf', 'prof',
    'locp', 'actp', 'etatp', 'senc', 'obs','obsm', 'choc', 'manv', 'motor',
    'jour', 'mois', 'lum', 'agg', 'int', 'atm', 'col', 'catr',
    'circ', 'vosp', 'prof', 'plan', 'surf', 'infra', 'situ']

# 3. Calcul des V de Cramer
res = {var: cramers_v(df[var], df['grav']) for var in vars_cat}

# 4. Passage en DataFrame et tri
cv_df = pd.DataFrame.from_dict(res, orient='index', columns=['V de Cramer'])
cv_df = cv_df.sort_values('V de Cramer', ascending=False)

# 5. Heatmap triée
plt.figure(figsize=(6, len(vars_cat)*0.5))
sns.heatmap(cv_df, annot=True, cmap='coolwarm', vmin=0, vmax=1, cbar=True)
plt.title("V de Cramer triés entre 'grav' et autres variables catégorielles")
plt.ylabel("Variables (triées)")
plt.tight_layout()
plt.show()
```

### Analyse des V de Cramer entre **grav** et variables catégorielles

- **`secu1` (ceinture de sécurité à l’avant)** : V ≈ 0,28  
- **`catv` (catégorie de véhicule)** : V ≈ 0,24  
- **`secu2` (ceinture arrière)** & **`locp` (localisation du choc)** : V ≈ 0,18  
- **`catu` (catégorie d’usager)**, **`actp` (type de manœuvre)** & **`agg` (agrégat)** : V ≈ 0,17  
- **`manv`, `obs`, `col`, `trajet`, `obsm`, `sexe`, `etatp`** : V ≈ 0,13–0,16  
- **Variables faiblement associées** (V < 0,05) :  
  - **Mois**, **Jour**, **Surface**, **Condition atmosphérique**, **Infra**, **Prof**, **Vitesse autorisée**, etc.

> **Conclusion**

### 3.2 Encodage des variables catégorielles

Les variables explicatives sont **catégorielles** (représentées souvent par des codes numériques dans les données brutes). Pour que les algorithmes de machine learning les exploitent correctement, il faut les **encoder** en variables numériques appropriées :

- Pour les algorithmes basés sur des arbres (Random Forest, XGBoost), il est possible de garder les codes numériques tels quels en s'assurant qu'ils sont bien considérés comme des catégories ordonnées arbitrairement. Cependant, il est plus sûr de réaliser un **one-hot encoding** (encodage binaire) afin de ne pas introduire d'ordre artificiel.
- Pour les algorithmes linéaires (Régression logistique) ou d'autres modèles, un encodage one-hot est généralement nécessaire..

Dans notre approche, nous allons **One-Hot encoder** toutes les variables catégorielles pertinentes :
